# Handling missing values

```python
df.dropna(subset=['Customer_ID'], inplace = True)
```

```python
df.shape
```

```python
df['Age'].unique()
```

```python
import re
def extract_age(age):
    age_num = re.findall('[0-9]+', str(age))
    if len(age_num) > 0 :
        return age_num[0]
    else :
        return age

df['Age'] = df['Age'].apply(lambda x : extract_age(x))
```

```python
df['Age'].unique()
```

```python
df_age = df[df['Age'] != 'nan years'] ['Age']
```

```python
df_age
```

```python
df_age.unique()
```

```python
df_age.dropna().astype('int64').median()
```

```python
age_median = int(df_age.dropna().astype('int64').median())
```

```python
age_median
```

```python
df.replace('nan years' , age_median, inplace = True)
```

```python
df['Age'] = df['Age'].fillna(age_median)
```

```python
df['Age'].unique()
```

```python
age_median = df['Age'].median()
df.loc[df['Age'] > 120, 'Age'] = age_median
```

```python
df['Age'].unique()
```

```python
df.isnull().sum()
```

```python
df['Purchase_Amount'] = df['Purchase_Amount'].fillna(df['Purchase_Amount'].median())
```

```python
df.isnull().sum()
```

```python
df['Feedback_Score'] = df['Feedback_Score'].fillna(df['Feedback_Score'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['City'] = df['City'].fillna(df['City'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['Country'] = df['Country'].fillna(df['Country'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['Last_Purchase_Date'] = df['Last_Purchase_Date'].ffill()
```

```python
df.isnull().sum()
```
